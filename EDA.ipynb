```python
#Importing the necessary librarires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Removing the duplicate values
df.drop_duplicates()
```

```python
#Drop the unwanted columns
df=df.drop(['video_id','title','tags','comment_count','comments_disabled','ratings_disabled','thumbnail_link','description'],axis=1)
```
